#matplot lib on pokemon data

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = pd.read_csv('Pokemon.csv')

file.head(10)
```

```python
data = file.head(10)
x = data.Name
y = data.Attack
y2 = data.HP
y3 = data.Defense


plt.title('Pokemon attack powers')
plt.xlabel('Pokemon name')
plt.ylabel('stats')


plt.plot(x, y, 'g^-', label = 'attack')
plt.plot(x, y2, 'r.-', label = 'HP')
plt.plot(x, y3, 'b*-', label = 'Defense')

plt.legend()
plt.xticks(rotation = 90)

plt.savefig('Pokemon_graph', dpi = 300)
plt.show()
```

```python
bins = [20,40,60,80,100,120,140,160,180,200]
plt.hist(file['Sp. Atk'], bins=bins, color = 'green')
plt.xticks(bins)
plt.xlabel('Special Attack Power')
plt.ylabel('Number of Pokemons')
plt.title('Number of pokemons with a certain level of attack power')

plt.show()
```

```python
plt.style.use('ggplot')

gen1 = file.loc[file.Generation == 1].count().iloc[0]
gen2 = file.loc[file.Generation == 2].count().iloc[0]
gen3 = file.loc[file.Generation == 3].count().iloc[0]
gen4 = file.loc[file.Generation == 4].count().iloc[0]
gen5 = file.loc[file.Generation == 5].count().iloc[0]
gen6 = file.loc[file.Generation == 6].count().iloc[0]

generations = [gen1, gen2, gen3, gen4, gen5, gen6]
labels = ['gen1', 'gen2', 'gen3', 'gen4', 'gen5', 'gen6']
explode = (.1,.1,.3,.1,.2,.4)
plt.title("Distribution of Pokemons in each generations")

plt.pie(generations, labels = labels, autopct='%.2f %%', explode = explode, pctdistance = 0.8)

plt.show()
```

```python
plt.style.use('default')

plt.figure(figsize = (7,8))

gen1 = file.loc[file.Generation == 1]['Total']
gen2 = file.loc[file.Generation == 2]['Total']
gen3 = file.loc[file.Generation == 3]['Total']
gen4 = file.loc[file.Generation == 4]['Total']
gen5 = file.loc[file.Generation == 5]['Total']
gen6 = file.loc[file.Generation == 6]['Total']

generations = [gen1, gen2, gen3, gen4, gen5, gen6]
labels = ['generation 1', 'generation 2', 'generation 3', 'generation 4', 'generation 5', 'generation 6']

box = plt.boxplot(generations, tick_labels = labels, patch_artist = True, medianprops = {'linewidth':2, 'color':'blue'})

plt.title('Comparison of generations of pokemons', fontdict ={ 'fontname': 'Comic Sans MS', 'fontsize' : 18})



plt.xticks(rotation = -90)

for box in box['boxes']:
    #set edge colour
    box.set(color = 'purple', linewidth = 2)

    #set facecolor
    box.set(facecolor = 'lightgrey')
    
plt.show()
```
